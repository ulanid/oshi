Завдання 2: Розв’язати квадратне рівняння $ax^2+bx + c = 0$.  Напишіть функцію,яка приймає параметри квадратного рівняння, а повертає одне рішення або декілька.

Наприклад:  


```
solve_quad(a=1, b=-2, c=1) -> 1  
solve_quad(a=1, b=0, c=-1) -> (-1, 1)  
solve_quad(a=0, b=2, c=-1) -> 0.5
```







````python
def solve_quad(a: float, b: float, c: float):
  
  D = b**2-4*a*c
    
  if D == 0:
      x = -b/(2*a)
      return x
    
  elif D > 0:
        x1 = (-b+D**0.5)/(2*a)
        x2 = (-b-D**0.5)/(2*a)
        return x1, x2
    
  else:
        return "∅"

a = float(input("Введіть значення a: "))
b = float(input("Введіть значення b: "))
c = float(input("Введіть значення c: "))

result = solve_quad(a, b, c)
print(result)
````

Завдання 3: Реалізація функції merge. На вхід функція приймає 2 відсортованих масиви - `a` і `b`. Функція повинна повернути один відсортований масив, який містить всі елементи масивів `a` і `b`.

Наприклад: `merge([1, 3, 3, 5], [3, 5, 5, 7]) -> [1, 3, 3, 3, 5, 5, 5, 7]`

````python
from typing import List

def merge(a: List[int], b: List[int]) -> List[int]:
    array_a = len(a)
    array_b = len(b)
    i = 0
    j = 0
    
    arr = []
      
    while i < array_a and j < array_b:
        if a[i] < b[j]:
            arr.append(a[i])
            i += 1
        else:
            arr.append(b[j])
            j += 1
    while i < array_a:
        arr.append(a[i])
        i += 1
        
    while j < array_b:
        arr.append(b[j])
        j += 1      
    return arr

a_input = input("Введіть елементи масиву 'a': ")
a = [int(x) for x in a_input.split()]
a.sort()

b_input = input("Введіть елементи масиву 'b': ")
b = [int(x) for x in b_input.split()]
b.sort()

arr = merge(a, b)
print("Відсортований масив:", arr)
````

Завдання 4: Реалізація алгоритму [MergeSort](https://ru.wikipedia.org/wiki/%D0%A1%D0%BE%D1%80%D1%82%D0%B8%D1%80%D0%BE%D0%B2%D0%BA%D0%B0_%D1%81%D0%BB%D0%B8%D1%8F%D0%BD%D0%B8%D0%B5%D0%BC)..

Наприклад: `MergeSort([3, 1, 3, 5]) -> [1, 3, 3, 5]`

````python
from typing import List

def MergeSort(L: List[int]) -> List[int]:
  if len(L) > 1:
    left_array = L[:len(L)//2]
    right_array = L[len(L)//2:]

    MergeSort(left_array)
    MergeSort(right_array)

    i = 0
    j = 0
    k = 0
    
    while i < len(left_array) and j < len(right_array):
      if left_array[i] < right_array[j]:
        L[k] = left_array[i]
        i += 1
      else:
        L[k] = right_array[j]
        j += 1
      k += 1
      
    while i < len(left_array):
      L[k] = left_array[i]
      i += 1
      k += 1
      
    while j < len(right_array):
      L[k] = right_array[j]
      j += 1
      k += 1      

input_array = input("Введіть елементи масиву: ")
arr_test = [int(x) for x in input_array.split()]  
MergeSort(arr_test)
print("Відсортований масив: ", arr_test)
````

Завдання 6: Написати функцію яка на вхід приймає 3 масиви – `X, Y, Z`. функція повертатиме `True` у випадку, коли вони рівні та `False` – якщо ні.

Рекомендація до завдання: Не використовуйте оператор `==`

````python
def is_equal(X, Y, Z):

    if X.shape != Y.shape or X.shape != Z.shape:
        return False
    
    return np.array_equal(X, Y) and np.array_equal(Y, Z)

X_input = input("Введіть елементи масиву 'X': ")
Y_input = input("Введіть елементи масиву 'Y': ")
Z_input = input("Введіть елементи масиву 'Z': ")

X = np.array([int(x) for x in X_input.split()])
Y = np.array([int(y) for y in Y_input.split()])
Z = np.array([int(z) for z in Z_input.split()])

print(X)
print(Y)
print(Z)

result = is_equal(X, Y, Z)
print(result)
````

Завдання 8: Написати функцію яка приймає на вхід двовимірний масив, а повертає масив з діагональними елементами цієї матриці.

Наприклад: `[[1, 2], [3, 4]] -> [1 4]`

````python
def get_diag(x):
    
    x_matrix = np.array(x)

    diagonal_elements = np.diag(x_matrix)
    
    return diagonal_elements

x_input = input("Введіть елементи двовимірного масиву розділяючи рядки за допомогою ';': ")

x = [[int(num) for num in row.split()] for row in x_input.split(';')]

print(x)
diagonal_elements = get_diag(x)
print("Діагональні елементи матриці:", diagonal_elements)

````

Завдання 9: Написати функцію яка приймає на вхід двомірний масив, а повертає масив елементів які знаходяться на другорядній діагоналі. Приклад: `[[1, 2], [3, 4]] -> [2 3]`

````python
def sec_diag(X):

    X_matrix = np.array(X)
    
    secondary_diagonal_elements = np.diag(np.fliplr(X_matrix))
    
    return secondary_diagonal_elements

X_input = input("Введіть елементи двовимірного масиву розділяючи рядки за допомогою ';': ")

X = [[int(num) for num in row.split()] for row in X_input.split(';')]

print(X)
secondary_diagonal_elements = sec_diag(X)
print("Елементи на другорядній діагоналі:", secondary_diagonal_elements)
````
